### 3.1 索引器

不建议将纯浮点或任何混合类型作为索引

#### 表的列索引

```python
demo_src = pd.read_csv('data/learn_pandas.csv', usecols=['School', 'Grade', 'Name', 'Gender', 'Weight', 'Transfer'])
# demo_src = demo[demo.columns[:7]]
demo_src['Name'].head()
demo_src.Name.head()    # 取出单列，与上面等价，列名中不含空格
```

#### 序列的行索引

```python
demo = pd.Series([1,2,3,4,5,6], index=['b','a','c','a','e','d'])
demo['a']   # 如果索引有多个值对应，返回一个Series
demo[['c', 'b']]
# 取出两个索引间的元素，且该索引具有唯一性，可以使用切片
demo['b':'e':2]   # 隔两步取一个数，索引包含两端

# 重复索引，排序后才可以切片
print(demo.sort_index())
demo.sort_index()['a':'b']

demo[1:3]   # 整数索引，不包含右端
```

#### loc索引器(基于元素)

loc[*, *], * 的位置可以为-单个元素，元素列表，元素切片，布尔列表，函数

```python
# print(demo_src)
df_demo = demo_src.set_index('Name')    # 将Name设置索引
df_demo.head()
```

```python
df_demo.loc['Mei Sun']
df_demo.loc['Mei Sun', 'Grade']

df_demo.loc[['Mei Sun', 'Xiaojuan Sun'], ['Grade', 'Transfer']]

# 整数索引，需保证索引
df_demo.shape   # (200,5)
df_loc_slice = df_demo.copy()
df_loc_slice.index = range(0, df_demo.shape[0]) # 设置整数索引
df_loc_slice.loc[2:15:2]
```

```python
df_demo.loc[df_demo.Grade.isin(['Freshman', 'Senior'])].head()
```

```python
con1 = df_demo.Grade=='Freshman'
con2 = df_demo.Grade=='Senior'
con = con1 | con2
df_demo.loc[con]
```

```python
df_demo.loc[(df_demo.Grade=='Freshman') | (df_demo.Grade=='Senior')]
```

```python
df_demo.loc[(df_demo.Gender=='Female') & (df_demo.Weight<50)].head()
df_demo.loc[(df_demo.Gender=='Female') & ~(df_demo.Weight<50)].head()
```

```python
df_demo.loc['Quan Zhao', 'Gender']
```

#### iloc针对位置进行索引

整数，整数列表，整数切片，布尔列表，函数

```python
df_demo.iloc[1:3, 2:4]
df_demo.iloc[1:4]
df_demo.iloc[lambda x: slice(1,4)]
```

```python
# 布尔筛选上的区分,iloc需转换为值 0或1
df_demo.loc[df_demo.Weight>80].head()
df_demo.iloc[(df_demo.Weight>80).values].head()
```

#### query方法

采用字符串表达式输入 query 查询数据，结果返回布尔列表

```python
# 传入字符串
df_demo.query('Grade == "Senior" & Weight>70')
df_demo.query('Weight > Weight.mean()').head()
# 对于含有空格的列名，使用 `col name`方式引用
```

```python
# query()注册的其他形式, and,or,in,not in
df_demo.query('Grade == "Senior" and Weight>70')

# 外部变量的引用@
thred = 70
df_demo.query('(Grade == "Senior") and (Weight>@thred)')
```

#### 随机抽样

行--样本
列--特征


sample(n, axis, frac, replace, weights)

抽样数量在，抽样方向(0行，1列)，抽样比例(0.3表示从总体中抽出30%)

replace:是否放回样本，replace=True表示有放回，weights:每个样本抽样相对概率

```python
df_sample = pd.DataFrame({'id': list('abcde'), 'value':[1,2,3,4,90]})
# 抽样数量3，以value列相对大小为抽样概率
print(df_sample.value)
df_sample.sample(3, replace=True, weights=df_sample.value)
```

#### IndexSlice对象
Slice对象有两种形式：

- loc[idx[*,*]]
- loc[idx[*,*], idx[*,*]]

```python
np.random.seed(0)
l1, l2 = list('ABC'),list('abc')
l3, l4 = list('DEF'),list('def')

# 多级索引
mul_index1 = pd.MultiIndex.from_product((l1,l2), names=('Upper', 'Lower'))
mul_index2 = pd.MultiIndex.from_product((l3,l4), names=('Big', 'Small'))

# 创建索引不重复的DataFrame
# 多级索引对应的规格数：每级索引数相乘(3×3)
df_demo = pd.DataFrame(np.random.randint(-9,10,(9,9)), index=mul_index1, columns=mul_index2)
df_demo
```

```python
# 定义并使用silce对象
idx = pd.IndexSlice
# loc[*,*]不能进行多层索引切片
df_demo.loc[idx[('C','b'):, ('D', 'e'):]]
df_demo.loc[idx[:('B', 'b'), lambda x:x.mean()>0]]
```

```python
# 分层切片
# 此时不支持使用函数
idx = pd.IndexSlice
df_demo.loc[idx[:'A', 'b':], idx['E':, 'e']]
```

### 3.3索引的常用方法

#### 索引层的交换和删除

- swaplevel交换两层
- reorder_levels交换任意层

```python
# 以3级索引为例2×2×2
np.random.seed(0)
L1,L2,L3 = ['A','B'],['a','b'],['alpha','beta']
mul_index1 = pd.MultiIndex.from_product([L1,L2,L3],names=('Upper','Lower','Extra'))
L4,L5,L6 = ['C','D'],['c','d'],['cat','dog']
mul_index2 = pd.MultiIndex.from_product([L4,L5,L6],names=('Big','Small','Other'))

df_demo = pd.DataFrame(np.random.randint(-9,10,(8,8)), index=mul_index1, columns=mul_index2)
df_demo
```

```python
# 交换两层,axis=1表示列索引，0列和2列交换(i,j,axis)
df_demo.swaplevel(0,2,axis=1).head()

# reorder_levels交换任意两层([order],axis),[2,0,1]为新的行索引排序
df_demo.reorder_levels([2,0,1], axis=0).head()
```

```python
# 删除某层索引
df_demo.droplevel(1, axis=1)
df_demo.droplevel([0,1], axis=1)    # 删除多层
df_demo.droplevel('Upper')
```

#### 索引属性的修改

```python
# rename_axis对索引层名字命名,一般为字典
df_demo.rename_axis(index={'Upper':'Changed_row'}, columns={'Other':'Change_animal'})

# rename修改索引值,指定对应的层号
df_demo.rename(columns={'cat':'Change_value'}, level=2).head()

# 把2层的行索引变为大写
df_demo.rename(index=(lambda x:str.upper(x)), level=2)

# inplace 为rename的参数
# inplace表示是否原地修改DataFrame而不创建副本，默认为False
inplace_demo = df_demo.rename_axis(index=(lambda x: 'Change_func' if x=='Upper' else x))
inplace_demo
```

```python
# 对所有元素的替代，使用迭代器
demo_inplace = iter(list('abcdefgh'))
df_demo.rename(index=lambda x:next(demo_inplace), level=2)
```

```python
# map函数，作用在index
# 同lambda类似，但是lambda传入的是标量，map传入的是数组
df_temp = df_demo.copy()
new_index = df_temp.index.map(lambda x:(x[0],x[1],str.upper(x[2])))
df_temp.index = new_index
df_temp
```

```python
# map另一个作用，对多级索引的压缩
df_temp = df_demo.copy()
zip_index = df_temp.index.map(lambda x:(x[0]+'-'+x[1]+'-'+x[2]))
df_temp.index = zip_index
df_temp.head()
```

```python
# 展开索引
unzip_index = df_temp.index.map(lambda x:tuple(x.split('-')))
df_temp.index = unzip_index
df_temp
```

#### 索引的设置与重置

```python
df_new = pd.DataFrame({'A':list('aacd'),'B':list('PQRT'),'C':[1,2,3,4]})
df_new
```

```python
# 设置索引，append表示是否保留原来的索引
df_new.set_index('A', append=True)
df_new.set_index(['A','B']) #指定多个索引

# 用Series自定义索引
df_index = pd.Series(list('qwer'),name='define')
df_new = df_new.set_index(df_index)

# 将索引还原回列
df_new.reset_index('define')
df_new.reset_index('define', drop=True) # 将索引直接丢弃，不还原
```

#### 一般索引运算

```python
df_set1 = pd.DataFrame([[0,1], [1,2], [3,4]], index = pd.Index(['a','b','a'], name='id1'))
df_set2 = pd.DataFrame([[4,5], [2,6], [7,1]], index = pd.Index(['b','b','c'], name='id2'))
df_set2
```

```python
# 对索引去重
id1,id2 = df_set1.index.unique(), df_set2.index.unique()
id1.intersection(id2)
id1.union(id2)
id1.difference(id2)
id1.symmetric_difference(id2)
```
